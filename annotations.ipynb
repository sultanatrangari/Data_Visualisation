Simple Title Annotations

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-4, 4, 1024)
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)

plt.title('A polynomial')
plt.plot(X, Y, c = 'K')
plt.show()
```

Labeling the X and Y axes

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-4, 4, 1024)
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)

plt.title('Graph Title')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.plot(X, Y, c = 'K')
plt.show()
```

Adding Text Anywhere

```python
import numpy as np 
import matplotlib.pyplot as plt

X = np.linspace(-4, 4, 1024) 
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)

plt.text(-0.3, -0.28, 'Nice Curve')

plt.plot(X, Y, c = 'k') 
plt.show() 
```

Bounded Box COntrol

```python
import numpy as np 
import matplotlib.pyplot as plt

X = np.linspace(-4, 4, 1024) 
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)

box = { 'facecolor' : '.75', 'edgecolor' : 'k', 'boxstyle' : 'round'}

plt.text(-0.5, -0.20, 'Nice Curve..in a Box', bbox = box)
plt.plot(X, Y, c = 'K')
plt.show()
```

Adding an Arrow to a chat

```python
import numpy as np 
import matplotlib.pyplot as plt

X = np.linspace(-4, 4, 1024) 
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)

plt.annotate('Low Point', ha = 'center', va = 'bottom', xytext = (-1.5, 3.), xy = (0.75, -2.7), arrowprops = {'facecolor' : 'black', 'shrink' : 0.05})
plt.plot(X, Y)
plt.show()
```

Adding a Grid to a Chart

```python
import numpy as np 
import matplotlib.pyplot as plt

X = np.linspace(-4, 4, 1024) 
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)

plt.plot(X, Y, c = 'k') 
plt.grid(True, lw = 2, ls = '--', c = '.75') 
plt.show() 
```

Adding Ticks to a Chart

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

X = np.linspace(-4, 4, 1024) 
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)

ax = plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))

plt.grid(True, which = 'both')
plt.plot(X, Y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

X = np.linspace(-15, 15, 1024) 
Y = np.sinc(X)

ax = plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))

plt.grid(True, which = 'both')
plt.plot(X, Y)
plt.show()
```

Labeling our Ticks

```python
import numpy as np 
import matplotlib.ticker as ticker 
import matplotlib.pyplot as plt

name_list = ('Mike', 'Brad', 'Stan', 'Jon', 'Matt') 
value_list = np.random.randint(0, 99, size = len(name_list)) 
pos_list = np.arange(len(name_list))

ax = plt.axes() 
ax.xaxis.set_major_locator(ticker.FixedLocator((pos_list))) 
ax.xaxis.set_major_formatter(ticker.FixedFormatter((name_list)))

plt.bar(pos_list, value_list, color = '.75', align = 'center') 
plt.show() 
```

Adding Ticks (The Easy Way)

```python
import numpy as np 
import matplotlib.pyplot as plt

name_list = ('Mike', 'Brad', 'Stan', 'Jon', 'Matt') 
value_list = np.random.randint(0, 99, size = len(name_list)) 
pos_list = np.arange(len(name_list))

plt.bar(pos_list, value_list, color = '.75', align = 'center')
plt.xticks(pos_list, name_list)
plt.show()
```
